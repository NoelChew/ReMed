```python
import folium
import pandas as pd
from ipywidgets import widgets, interact
from datetime import datetime
from dateutil.relativedelta import relativedelta



monthOffset = 1
medicineType = 1

# Define function to bind value of the input to the output variable 
def f(x):
    global monthOffset
    monthOffset = x
    #output_slider_variable.value = str(x)
    #print (output_slider_variable.value)
    # Create and output new int variable with value of slider

print ("Months left Expiring")
# Create input slider with default value = 10  
interact(f, x=widgets.IntSlider(min=1,max=12,step=1,value=2));
#interact(f, x=10)


#new_variable = int(output_slider_variable.value)
#print new_variable


drugs={'Ibuprofen':'Ibuprofen','Paracetamol':'Paracetamol'}

def select_drug(drug):
    global medicineType
    if drug == 'Paracetamol':
        medicineType = 2
    else:
        medicineType = 1
    print ("Selected " + str(drugs[drug]) + " Expiring in " + str(monthOffset) + " Months.")
    

scW = widgets.Select(options=drugs.keys())
#init = scW.value
widgets.interactive(select_drug, drug=scW)

```

```python

#3.154205, 101.713112
map = folium.Map(location=[3.154205, 101.713112],
                   zoom_start=13)

medicine = pd.read_csv("data/remed2.csv")


marker_cluster = folium.MarkerCluster("Medicine Cluster").add_to(map)

for each in medicine.iterrows():
    global medicineType
    date_object = datetime.strptime(each[1]['date_time'], '%b %Y')
    if medicineType == (each[1]['medicineType']):
        if date_object < datetime.today()+ relativedelta(months=monthOffset):
            folium.Marker([each[1]['lat'],each[1]['lng']], popup="Medicine Cluster").add_to(marker_cluster)


map
```
